```python
import json
```

```python
allQcs = {}
for qc in open('./ground_res_with_target.jsonl', 'r'):
    obj = json.loads(qc)
    if obj['id'] in allQcs:
        print('duplicate id', obj['id'])
    allQcs[obj['id']] = obj['res']

allQcs
```

```python
missing = {}
count = 0
for qc in open('./ground_res_without_target.jsonl', 'r'):
    obj = json.loads(qc)
    for concept in obj['res']['qc']:
        if concept not in allQcs[obj['id']]['qc']:
            if obj['id'] not in missing:
                missing[obj['id']] = []
            missing[obj['id']].append(concept)
            count += 1

count, missing
```
