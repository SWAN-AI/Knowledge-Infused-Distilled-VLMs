```python
import json
from tqdm import tqdm
```

```python
fileopen = open('/scratch/rahul.garg/noTarget_Rel_Cmgs_h1.jsonl', 'r')
outputFile = open('/scratch/rahul.garg/noTarget_Rel_Cmgs_h1_noEmbed.jsonl', 'w')
for line in tqdm(fileopen, total=12000):
    #dict_keys(['class', 'nodeDataArray', 'linkDataArray', 'contextEmbed'])
    # print(json.loads(line).keys())
    # print(json.loads(line)['graph'].keys())
    obj = json.loads(line)
    line2 = {
        'id': obj['id'],
        'graph': {
            'class': obj['graph']['class'],
            'nodeDataArray': obj['graph']['nodeDataArray'],
            'linkDataArray': obj['graph']['linkDataArray']
        }
    }
    outputFile.write(json.dumps(line2) + '\n')
outputFile.close()
fileopen.close()
```
