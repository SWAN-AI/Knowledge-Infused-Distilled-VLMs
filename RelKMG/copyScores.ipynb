```python
import json
from tqdm import tqdm
```

```python
inputFile = '/scratch/rahul.garg/noTarget_Rel_Cmgs_h1_noEmbed.jsonl'
inputfile2 = '/scratch/rahul.garg/conceptNet/allCmgs_noTargetFixed_h1.jsonl'
outputFile = '/scratch/rahul.garg/conceptNet/allCmgs_noTargetFixed_h1_w_scores.jsonl'

inputFile = open(inputFile, 'r')
# print first line
mainInputLines = inputFile.readlines()
inputFile.close()
scores = {}
for line in tqdm(mainInputLines):
    obj = json.loads(line)
    if obj['id'] not in scores:
        scores[obj['id']] = {}

    for node in obj['graph']['nodeDataArray']:
        scores[obj['id']][node['key']] = node['score']

```

```python
input2File = open(inputfile2, 'r')
# add the scores to this, this has subset of nodes
input2Lines = input2File.readlines()
input2File.close()
outputFile = open(outputFile, 'w')
for line in tqdm(input2Lines):
    obj = json.loads(line)
    if obj['id'] in scores:
        for node in obj['graph']['nodeDataArray']:
            if node['key'] in scores[obj['id']]:
                node['score'] = scores[obj['id']][node['key']]
    else:
        print(f"Could not find {obj['id']} in scores")
        break
            
    outputFile.write(json.dumps(obj) + '\n')
```
